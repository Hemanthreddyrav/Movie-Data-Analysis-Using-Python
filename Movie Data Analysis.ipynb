```python
import numpy as np # For array processing
import pandas as pd # For data analytics
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
movie_df = pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')
movie_df
# Reading the csv file into jupiter using pd.read_csv
```

```python
import json 
# This allows you to load JSON(Java script object Notation) data from files into usable Python dicts, and alsoconvert your Python objects to JSON format.
# It is difficult to manipulate and clean data and filtering data in json format 
```

```python
movie_data = movie_df.copy()
# Makes a copy of movie_df called movie_data to modify without affecting the original.
```

```python
columns_with_dicts = ['genres','keywords','production_companies','production_countries','spoken_languages']
# Creates a list of columns that contain dictionary data in JSON format.
```

```python
for column in columns_with_dicts:
    movie_data[column] =  movie_data[column].apply(lambda x: [item['name'] for item in json.loads(x)])

# Loops through the columns containing JSON dicts.
# Applies a lambda to parse the JSON into a dict using json.loads()
# Extracts the 'name' value into a list using a list comprehension.
# Saves the list back into the DataFrame column.
```

```python
movie_data
```

```python
columns_with_lists = ['genres','keywords','production_companies','production_countries','spoken_languages']
# Creates a list of columns that contain list data
```

```python
for column in columns_with_lists:
    movie_data[column] = movie_data[column].apply(lambda x:x[0] if x else None)
    
# Loops through list columns
# Lambda returns first element or None for missing values
```

```python
# Numerical columns int,float

movie_data['runtime'].fillna(movie_data['runtime'].mean(),inplace=True)

# Fills NaN in runtime column with mean
```

```python
# Categorical columns object

movie_data['genres'].fillna(movie_data['genres'].mode()[0],inplace=True)
movie_data['homepage'].fillna(movie_data['homepage'].mode()[0],inplace=True)
movie_data['keywords'].fillna(movie_data['keywords'].mode()[0],inplace=True)
movie_data['production_companies'].fillna(movie_data['production_companies'].mode()[0],inplace=True)
movie_data['production_countries'].fillna(movie_data['production_countries'].mode()[0],inplace=True)
movie_data['spoken_languages'].fillna(movie_data['spoken_languages'].mode()[0],inplace=True)

# Fills NaN in genres,homepage,keywords,production_companies,production_countries,spoken_languages with mode
```

```python
# Forward fill data imputation
movie_data['release_date']=movie_data['release_date'].fillna(method='ffill')

# using fillna(method='ffill') forward fills NaN values in release_date column
```

```python
movie_data['overview'].fillna('No overview available',inplace=True)
movie_data['tagline'].fillna('No tagline available',inplace=True)

```

```python
plt.figure(figsize=(8,5))
sb.regplot(data=movie_data,x='budget',y='popularity',scatter=True)
plt.xlabel('Budget')
plt.ylabel('Popularity')
plt.show()
# regplot gives us a linear regression line
```

```python
# The budget and popularity regplot() shows us that there is a positive correlation between them.
# Its means that the movies with higher budjet tends to be more popular than the movies with low bubjet.
# However there are some movies where the budjet is low but the populatity is the highest this maybe due to the movie cast or genre.
# It means that although there is a positive relation between the budget of the movie and popularity it is not a perfect relation.
# Other factors can also contribute to the popularity.
 
```

```python
Top_25_production_companies = pd.pivot_table(data=movie_data,index='production_companies',values='title',
                                             aggfunc='count').sort_values(by='title',ascending=False).head(25)
Top_25_production_companies = Top_25_production_companies.rename(columns={'title': 'Number_of_movies'})
Top_25_production_companies

# Top_25_production_companies, contains information about the top 25 production companies based on the number of movies they have produced
```

```python
Top_500_movies = movie_data[['budget','revenue','runtime']].head(500).sort_values(by='revenue',ascending=False)
Top_500_movies
# Filters for top 500 by revenue
```

```python
Top_500_movies.describe()

# Central tendencies of budget,revenue and runtime
```

```python
plt.figure(figsize=(10,5))
sb.boxplot(data=Top_500_movies,orient='h')
plt.xscale('log')
plt.show()
# the box and whisker plot for budget vs revenue shows that the company is generating more revenue than it is spending. This is a good thing, but it is important to make sure that the company is budgeting appropriately.
```
